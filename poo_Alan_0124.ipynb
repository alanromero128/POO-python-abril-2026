### Conceptos de POO (Programacion orientada a objetos) en python

### El promp en IA gemini
lenguaje Python, crear la clase "animal" con los atributos "nombre" , "raza" crear el conductor animal, y las funciones ladrar(),correr(),crear los objetos pitbull, Chihuahua, nivel principiantes, no es interactivo

```python
# Definición de la clase
class Animal:
    # El constructor (__init__) inicializa los atributos del objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función o método para ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau!🔺")

    # Función o método para correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente.⚽")

    def Morder( self):
        print(f"{self.nombre} está mordiendo a la vecina🏅")

# Creación de los objetos (Instancias)

# 1. Objeto Pitbull
pitbull = Animal("Zeus", "Pitbull")

# 2. Objeto Chihuahua
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino","chafa")
# Uso de los objetos
print(f"Perro 1: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

print(f"Perro 2: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()
print("-" * 20)
print(f"Perro 3: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()
print("-" * 20)
print("Programa creado por Alan Romero  NL = 47")
```
